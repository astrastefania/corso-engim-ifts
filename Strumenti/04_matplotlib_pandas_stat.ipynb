# Matplotlib (seconda parte)

In questo notebook andremo ad utilizzare Scikit-learn (libreria di Machine Learning http://scikit-learn.org), solo per scaricare il dataset che importeremo come Pandas dataframe.

Vedremo i principali grafici statistici menzionati a partire su un Pandas dataframe.

```python
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data = load_iris()

iris = pd.DataFrame(data.data, columns=data.feature_names)
iris.head(10)
```

```python
iris.columns
```

## Istogrammi

```python
iris['sepal length (cm)'].plot('hist')
plt.title('Istogramma della largezza dei sepali')

plt.show()
```

```python
iris['petal width (cm)'].plot('hist')
plt.title('Istogramma della largezza dei petali')

plt.show()
```

```python
iris.hist()

plt.show()
```

## Boxplot (diagramma a scatola e baffi)

```python
iris.boxplot()

plt.show()
```

```python
iris.boxplot(column='sepal length (cm)')
plt.show()
```

```python
iris.plot(y='petal length (cm)', kind='box')
plt.show()
```

## Scatter plot (grafico a dispersione)

```python
iris.plot(x='sepal length (cm)', y='sepal width (cm)', kind='scatter')
plt.title('Grafico a disperzione della larghezza sulla lunghezza dei sepali')

plt.show()
```

## CDF (Cumulative Distribution Function)

```python
# Costruiamo una CDF 
iris['sepal length (cm)'].plot(kind='hist', bins=25, normed=True, cumulative=True)
plt.title('CDF della lunghezza dei sepali')

plt.show()
```

### Grafico a dispersione dell'entrate settimanali in funzione dell'et

```python
group_age = survey.groupby('Age').mean()
```

```python
group_age.head()
```

Usiamo **`plt.scatter( )`** passandogli i valori da mettere sul grafico.  
Ricordiamo che per richiamare gli indici usiamo l'attributo **`.index`** sul nostro dataframe.  

Per maggiori informazioni sugli argomenti possiamo eseguire **`help(plt.scatter)`**

```python
group_age.index
```

```python
plt.scatter(x=group_age.index, y=group_age['Weekly Earnings'])
plt.title("Grafico a dispersione della media delle entrate settimanali in funzione dell'et")

plt.show()
```

### Personalizziamo il grafico

**Grandezza grafico**
La impostiamo con **`plt.figure(figsize=(8,6))`**. 
Primo argomento di figsize  la dimensione dell'asse x, secondo argomento le dimensioni dell'asse y.

**Colorare punti del grafico in funzione di una variabile**
Usiamo l'argomento **c** per colorare i nostri punti in fuzione di una variabile categorica (per la colonna survey['Age'] ogni et viene letta come una categoria).

**Variare la trasparenza/opacit dei punti**
Giochiamo con l'argomento **alpha** per aumentare la trasparenza (ossia diminuire l'opacit) dei punti sul grafico.
Questo  molto utile nel caso si molti dati sovrapprosti

```python
plt.figure(figsize=(8,6))

plt.scatter(x=survey.Age, y=survey['Weekly Earnings'], c=survey['Age'], alpha=0.04)
plt.title("Grafico a dispersione delle entrate settimanali in funzione dell'et")


plt.show()
```

```python
plt.figure(figsize=(8,6))


plt.scatter(x=survey.Age, y=survey['Food & Drink Prep'], c=survey['Age'], alpha=0.07)
plt.title("Grafico a dispersione del tempo per la preparazine del cibo in funzione dell'et")

plt.show()
```

```python
plt.figure(figsize=(8,6))

plt.scatter(x=survey.Age, y=survey['Job Searching'], c=survey['Age'], alpha=0.3)
plt.title("Grafico a dispersione del tempo per la preparazine del cibo in funzione dell'et")

plt.show()
```
